**cardiffnlp/twitter-roberta-base-topic-latest**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_dataset
from nltk.translate.bleu_score import sentence_bleu
import torch
from nltk.translate.bleu_score import SmoothingFunction

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-topic-latest")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-topic-latest")

# Define the data processor class
class TweetTopicDataProcessor:
    def __init__(self, labels=None, labels_path=None):
        if not labels:
            self.topic_labels = (
                'arts_&_culture',
                'business_&_entrepreneurs',
                'celebrity_&_pop_culture',
                'diaries_&_daily_life',
                'family',
                'fashion_&_style',
                'film_tv_&_video',
                'fitness_&_health',
                'food_&_dining',
                'gaming',
                'learning_&_educational',
                'music',
                'news_&_social_concern',
                'other_hobbies',
                'relationships',
                'science_&_technology',
                'sports',
                'travel_&_adventure',
                'youth_&_student_life',
            )
            labels = self.topic_labels
        else:
            self.topic_labels = labels

    def convert_data(self, data):
        topics = ', '.join([topic[0] for topic in list(zip(self.topic_labels, data['gold_label_list'])) if topic[1] == 1])
        text_a = data['text']
        return {"text": text_a, "topics": topics}

    def get_examples(self, data_dir='cardiffnlp/super_tweeteval', split='train'):
        if split == "valid" or split == "dev":
            split = "validation"

        if data_dir is None:
            return []

        ds = load_dataset(data_dir, 'tweet_topic', split=split)
        return [self.convert_data(data) for data in ds]

# Load and preprocess dataset
processor = TweetTopicDataProcessor()
dataset = processor.get_examples(split="test")

# Helper function to predict topics
def predict_topics(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits.detach().numpy()
    predictions = torch.sigmoid(torch.tensor(logits)).numpy() > 0.5
    predicted_labels = [processor.topic_labels[i] for i in range(len(predictions[0])) if predictions[0][i]]
    return predicted_labels

# Evaluate the model using sentence_bleu
smoothing = SmoothingFunction().method1
scores = []
for example in dataset:
    predicted_topics = predict_topics(example["text"])
    reference_topics = example["topics"].split(', ')
    score = sentence_bleu([reference_topics], predicted_topics, weights=(1.0, 0, 0, 0), smoothing_function=smoothing)  # unigram BLEU with smoothing
    scores.append(score)

# Calculate and print average BLEU score
average_bleu = sum(scores) / len(scores)
print(f"Average BLEU score: {average_bleu:.4f}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_dataset
import torch
from sklearn.metrics import accuracy_score, f1_score
import numpy as np
from transformers import InputExample, DataProcessor

# Define the data processor for single-label classification
class TweetTopicDataProcessor:
    def __init__(self, labels=None, labels_path=None):
        if not labels:
            self.topic_labels = (
                'arts_&_culture',
                'business_&_entrepreneurs',
                'celebrity_&_pop_culture',
                'diaries_&_daily_life',
                'family',
                'fashion_&_style',
                'film_tv_&_video',
                'fitness_&_health',
                'food_&_dining',
                'gaming',
                'learning_&_educational',
                'music',
                'news_&_social_concern',
                'other_hobbies',
                'relationships',
                'science_&_technology',
                'sports',
                'travel_&_adventure',
                'youth_&_student_life',
            )
            labels = self.topic_labels
        else:
            self.topic_labels = labels

        # No need for `super().__init__()` here; initialize variables directly.
        self.labels = labels
        self.labels_path = labels_path

    def convert_data(self, data):
        # Select one active label (default to "other_hobbies" if none)
        topics = [topic for topic, is_present in zip(self.topic_labels, data['gold_label_list']) if is_present == 1]
        label = topics[0] if topics else "other_hobbies"

        text_a = data['text']

        return InputExample(
            text_a=text_a,
            label=self.topic_labels.index(label)  # Convert to index for classification
        )

    def get_examples(self, data_dir='cardiffnlp/super_tweeteval', split='train'):
        if split == "valid" or split == "dev":
            split = "validation"

        if data_dir is None:
            return []

        ds = load_dataset(data_dir, 'tweet_topic', split=split)
        return list(map(self.convert_data, ds))


    def convert_data(self, data):
        # The assumption is that one label is '1' (active), and the rest are '0'.
        topics = [topic for topic, is_present in zip(self.topic_labels, data['gold_label_list']) if is_present == 1]
        label = topics[0] if topics else "other_hobbies"  # Default to a category if none are present

        text_a = data['text']

        # Returning an InputExample where label is the index of the category
        return InputExample(
            guid=str(data.get('id', '')),  # Use the 'id' field if available, else an empty string
            text_a=text_a,
            label=self.topic_labels.index(label)  # Convert to index for classification
        )

    def get_examples(self, data_dir='cardiffnlp/super_tweeteval', split='train'):
        if split == "valid" or split == "dev":
            split = "validation"

        if data_dir is None:
            return []

        ds = load_dataset(data_dir, 'tweet_topic', split=split)
        return list(map(self.convert_data, ds))

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-topic-latest")
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-topic-latest", num_labels=19)  # 19 labels for single-label classification  # 20 labels for single-label classification

# Load and preprocess dataset
processor = TweetTopicDataProcessor()
dataset = processor.get_examples(split="test")

# Helper function to predict topics (single-label classification)
def predict_topic(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    predicted_label_idx = torch.argmax(logits, dim=-1).item()  # Get the index of the max logit
    predicted_label = processor.topic_labels[predicted_label_idx]
    return predicted_label

# Prepare the true labels and predicted labels
y_true = []
y_pred = []

print("Sample Predictions:\n")

for i, example in enumerate(dataset):
    predicted_label = predict_topic(example.text_a)
    reference_label = processor.topic_labels[example.label]

    y_true.append(reference_label)
    y_pred.append(predicted_label)

    if i < 5:  # Print the first 5 predictions
        print(f"Text: {example.text_a}")
        print(f"Predicted Label: {predicted_label}")
        print(f"True Label: {reference_label}\n")

for example in dataset:
    predicted_label = predict_topic(example.text_a)
    reference_label = processor.topic_labels[example.label]

    y_true.append(reference_label)
    y_pred.append(predicted_label)

# Calculate Accuracy, Macro F1, and Micro F1
accuracy = accuracy_score(y_true, y_pred)
macro_f1 = f1_score(y_true, y_pred, average='macro')
micro_f1 = f1_score(y_true, y_pred, average='micro')

print(f"Accuracy: {accuracy:.4f}")
print(f"Macro F1: {macro_f1:.4f}")
print(f"Micro F1: {micro_f1:.4f}")
```
